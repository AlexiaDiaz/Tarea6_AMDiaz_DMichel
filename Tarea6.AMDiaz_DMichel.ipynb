8. Demostrar la validez de la siguiente ecuación
$$p_{i+1}=\frac{n-i}{i+1}\frac{p}{1-p} p_i$$

**Código de solución estudiante 1**

$$\frac{p_{i+1}}{p_i} = \frac{n! p^{i+1}(1-p)^{n-i-1}}{(i+1)! (n-i-1)! }\frac{i!(n-i)!}{n! p^{i} (1-p)^{n-i}}$$

Resolver términos: $$ n!\rightarrow  n! $$ 

$$ \frac{i!}{i+1!} \rightarrow  \frac{1}{i+1}$$

$$\frac{n-i!}{n-i-1!} \rightarrow \frac{n-i}{1}$$

$$\frac{(1-p)^{n-i-1}}{(1-p)^{n-1}} \rightarrow \frac{1}{(1-p)}$$

$$\frac{p^{i+1}}{p^{i}} \rightarrow \frac{p}{1}$$


Al resolver los términos y sustituirlos por su resultado, tenemos:
$$\frac{p_{i+1}}{p_i} = \frac{(n-i)p}{ (i+1)(1-p)}$$  

Solamente queda dividir en 2 fracciones el numerador y denominador, así como pasar a $p_i$ hacia el otro lado para demostrar la validez de la ecuación:  

$$p_{i+1}=\frac{n-i}{i+1}\frac{p}{1-p} p_i$$


**Código de solución estudiante 2**


Se sabe que:$$p_{i+1}=\frac{n-i}{i+1}\frac{p}{1-p} p_i $$por lo que despejando:$$\frac{p_{i+1}}{p_i}=\frac{n-i}{i+1}\frac{p}{1-p} $$Y por propieddes factoriales puedo sustituir:$$\frac{(n+1)!}{n!}=n+1 ; \frac{n!}{(n+1)!}=\frac{1}{n+1}$$y:$$ (n-1)! = \frac{n!}{n} ; n!=n(1+n)!$$Y tomando en cuenta que:$$X\sim binomial(n,p) \longrightarrow p_i=P(X=i)=\frac{n!}{i!(n-i)!}p^i(1-p)^{n-i},\quad i=0,1,\cdots,n$$Entonces simplificando obtenemos::$$\frac{p_{i+1}}{p_i} = \frac{p(n-i)!(-p+i)(i+1)^{n+1-i}}{(i+1)(n+1-i)(-p+1)^{n-i}}$$

$$p_{i+1} = \frac{p(n-i)!(-p+i)(i+1)^{n+1-i}}{(i+1)(n+1-i)(-p+1)^{n-i}}p_i$$$$p_{i+1} = \frac{n-i}{i+1}\frac{p}{1-p}p_i$$


### Ejercicio 9:
9. Seguir un procedimiento similar al mostrado cuando se generó una distribución binomial, pero en esta caso genere un código que genere variables aletorias Poisson cuya función de distribución de probabilidad esta dada por:
$$P(k,\lambda)=\frac{e^{-\lambda}(\lambda)^k}{k!}$$
    

Demuestre matemáticamente que 
 $$P(k+1)=\frac{\lambda}{k+1}P(k)$$
 y a partir de esta relación genere variables aletorias que distribuyen poisson usando el método de la transformada inversa.

**Código de solución estudiante 1**

```python
def Gen_distr_discreta(U,p_acum):
    
    '''
    Tener en cuenta que este arreglo cuenta números empezando del 0
    
    Parámetros
    ----------
    U: "vector de números aleatorios"
    p_acum: "P.Acumulada de la distribución a generar"
    
    '''
    v = np.array(list(map(lambda j:sum(1 for i in p_acum if i<U[j]),range(len(U)))))
    return v
```

```python
def plot_histogram_discrete(distribucion,
                            label,
                            densidad=True):
    '''
    Parámetros
    ----------
    distribución: "señal de varibles aleatorias de un distribución DISCRETA dada"
    label: "label del legend a aparecer en el gráfica"
    densidad: "por defecto regresa el histograma en densidad"
    '''
    plt.figure(figsize=[8,4])
    y,x = np.histogram(distribucion,bins = len(set(distribucion)),density = densidad) 
    plt.bar(x[1:],y,label=label)
    plt.legend()
    plt.show()
```

```python
def poisson(n,p):
    Pr = np.zeros(n)
    Pr[0] = np.exp(-p)
    def pr(i):
        nonlocal Pr
        Pr[i+1]=(p/(i+1))*Pr[i]

    # Lleno el vector Pr usando compresión de listas
    [pr(i) for i in range(n-1)]
    return np.cumsum(Pr)
```

```python
# Declaración de parámetros
n=20
p=10
p_acum = poisson(n,p)
u = np.random.rand(10000)

# Usando el método de la transformada inversa
d_binomial = Gen_distr_discreta(u,p_acum)
plot_histogram_discrete(d_binomial,'usando función creada')
```

Solamente queda pasar el denominador hacia el otro lado para demostrar la igualación:

$$P(k+1)=\frac{\lambda}{k+1}P(k)$$

si:$$P(k+1) = \frac{\lambda}{K+1}P(k)$$Entonces:$$\frac{P(k+1)}{P(k)} = \frac{\lambda}{k+1}$$Usando la fórmula recursiva para calcular $P(k+1)$ en términos de $P(k)$:$$\frac{P(k+1)}{P(k)} = \frac{\frac{e^{-\lambda}(\lambda)^{k+1}}{(k+1)!}}{\frac{e^{-\lambda}(\lambda)^k}{k!}}$$También se hace uso de la propiedad$$\frac{k!}{(k+1)!}=\frac{1}{k+1}$$$$\frac{P(k+1)}{P(k)} = \frac{\lambda^{k+1}k!}{\lambda^{k}(k+1)!} = \frac{\lambda}{k+1}$$Por lo que:$$P(k+1) = \frac{\lambda}{k+1}P(k)$$


```python
def Gen_distr_discreta(p_acum: 'P.Acumulada de la distribución a generar',
                       indices: 'valores reales a generar aleatoriamente',
                       N: 'cantidad de números aleatorios a generar'):
    
    U =np.random.rand(N)
    # Diccionario de valores aleatorios
    rand2reales = {i: idx for i, idx in enumerate(indices)}

    # Series de los valores aletorios
    y = pd.Series([sum([1 for p in p_acum if p < ui]) for ui in U]).map(rand2reales)

    return y
def plot_histogram_discrete(distribucion:'señal de varibles aleatorias de un distribución DISCRETA dada',
                            label:'label del legend a aparecer en el gráfica',
                            densidad:'por defecto regresa el histograma en densidad'=True):
    # len(set(distribucion)) cuenta la cantidad de elementos distintos de la variable 'distribucion'
    plt.figure(figsize=[8,4])
    y, x = np.histogram(distribucion, bins=len(set(distribucion)), density=densidad) 
    plt.bar(x[1:], y, label=label)
    plt.legend()
    plt.grid()
    plt.title('Poisson',fontsize=14)
    plt.xlabel('Rango',fontsize=14)
    plt.ylabel('Probabilidad',fontsize=14)
    plt.show()  
# Función para la distribución acumulada de Poisson
#hecha como la P_acum_Binomial_o de la clase 10:
def poisson(l,k):
    z=np.zeros(k+1)
    z[0]=np.exp(-l)
    def z1(i):
        nonlocal z
        z[i+1]=z[i]*(l/(i+1))
    # Lleno el vector Pr usando compresión de listas
    [z1(i) for i in range(k-1)]
    return np.cumsum(z)
N = 10000;k = 10
Poisson=poisson(5,5)
# Usando el método de la transformada inversa
pl= Gen_distr_discreta(Poisson,np.arange(0,k+1),N)    
plot_histogram_discrete(pl,"Distribución Poisson")
```
